```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# we start by assigning the different values to different variables
# we take a value of gamma of 1 as we don't have any precise information about it

mu_1=0.08
mu_2=0.12
b_1=0.02
b_2=0.02
gamma=1
R_f=0.03

#we compute a function which plots the frontiers

def frontiers(sigma_1,sigma_2,rho,l,a): #returns the four different points
    
    X_1L1=(gamma*sigma_1*(1-rho**2))**(-1)*( (mu_1-R_f-b_1)/(sigma_1)- ((mu_2-R_f-b_2)/sigma_2)*rho)
    X_1L2=(gamma*sigma_1*(1-rho**2))**(-1)*( (mu_1-R_f-b_1)/(sigma_1)- ((mu_2-R_f+b_2)/sigma_2)*rho)
    
    X_1H1=(gamma*sigma_1*(1-rho**2))**(-1)*( (mu_1-R_f+b_1)/(sigma_1)- ((mu_2-R_f-b_2)/sigma_2)*rho)
    X_1H2=(gamma*sigma_1*(1-rho**2))**(-1)*( (mu_1-R_f+b_1)/(sigma_1)- ((mu_2-R_f+b_2)/sigma_2)*rho)
    
    X_2L1=(gamma*sigma_2*(1-rho**2))**(-1)*( (mu_2-R_f-b_2)/(sigma_2)- ((mu_1-R_f-b_1)/sigma_1)*rho)
    X_2L2=(gamma*sigma_2*(1-rho**2))**(-1)*( (mu_2-R_f+b_2)/(sigma_2)- ((mu_1-R_f-b_1)/sigma_1)*rho)
    
    X_2H1=(gamma*sigma_2*(1-rho**2))**(-1)*( (mu_2-R_f-b_2)/(sigma_2)- ((mu_1-R_f+b_1)/sigma_1)*rho)
    X_2H2=(gamma*sigma_2*(1-rho**2))**(-1)*( (mu_2-R_f+b_2)/(sigma_2)- ((mu_1-R_f+b_1)/sigma_1)*rho)
    
    abscisses=[X_1L1,X_1L2,X_1H1,X_1H2]
    ordonnees=[X_2L1,X_2L2,X_2H1,X_2H2]
    
    #shape of the no trading zone
    
    line_1_x_values=[X_1L1,X_1L2]
    line_1_y_values=[X_2L1,X_2L2]
    
    line_2_x_values=[X_1L2,X_1H2]
    line_2_y_values=[X_2L2,X_2H2]
    
    line_3_x_values=[X_1H1,X_1H2]
    line_3_y_values=[X_2H1,X_2H2]
    
    line_4_x_values=[X_1L1,X_1H1]
    line_4_y_values=[X_2L1,X_2H1]
    
    # boundaries between the buy/sell zone
    
    boundary_1_x_values=[X_1L1,X_1L1]
    boundary_1_y_values=[X_2L1,X_2L1-l]
    
    boundary_2_x_values=[X_1L1,X_1L1-l]
    boundary_2_y_values=[X_2L1,X_2L1]
    
    boundary_3_x_values=[X_1H1,X_1H1]
    boundary_3_y_values=[X_2H1,X_2H1-l]
    
    boundary_4_x_values=[X_1H1,X_1H1+l]
    boundary_4_y_values=[X_2H1,X_2H1]
    
    boundary_5_x_values=[X_1L2,X_1L2]
    boundary_5_y_values=[X_2L2,X_2L2+l]
    
    boundary_6_x_values=[X_1L2,X_1L2-l]
    boundary_6_y_values=[X_2L2,X_2L2]
    
    boundary_7_x_values=[X_1H2,X_1H2+l]
    boundary_7_y_values=[X_2H2,X_2H2]

    boundary_8_x_values=[X_1H2,X_1H2]
    boundary_8_y_values=[X_2H2,X_2H2+l]
        
    
    if a!=0:
        fig = plt.figure()
        ax = fig.add_subplot()
        plt.scatter(abscisses,ordonnees)
        plt.plot(line_1_x_values,line_1_y_values,color='black')
        plt.plot(line_2_x_values,line_2_y_values,color='black')
        plt.plot(line_3_x_values,line_3_y_values,color='black')
        plt.plot(line_4_x_values,line_4_y_values,color='black')
        plt.plot(boundary_1_x_values,boundary_1_y_values,color='black')
        plt.plot(boundary_2_x_values,boundary_2_y_values,color='black')
        plt.plot(boundary_3_x_values,boundary_3_y_values,color='black')
        plt.plot(boundary_4_x_values,boundary_4_y_values,color='black')
        plt.plot(boundary_5_x_values,boundary_5_y_values,color='black')
        plt.plot(boundary_6_x_values,boundary_6_y_values,color='black')
        plt.plot(boundary_7_x_values,boundary_7_y_values,color='black')
        plt.plot(boundary_8_x_values,boundary_8_y_values,color='black')
    
        ax.annotate('No Trade',xy=(X_1L1+a,X_2L1+a))
        ax.annotate('Buy/Buy',xy=(X_1L1-3*a,X_2L1-2*a))
        ax.annotate('No Trade/Buy',xy=(X_1L1+a,X_2L1-2*a))
        ax.annotate('Buy/No trade',xy=(X_1L1-5*a,X_2L1+a))
        ax.annotate('Sell/Buy',xy=(X_1H1+0.5*a,X_2H1-2*a))
        ax.annotate('Sell/No trade',xy=(X_1H1+a,X_2H1+a))
        ax.annotate('No trade/Sell',xy=(X_1L2+a,X_2L2+a))
        ax.annotate('Buy/Sell',xy=(X_1L2-2.5*a,X_2L2+a))
        ax.annotate('Sell/Sell',xy=(X_1H2+a,X_2H2+a))
    


        plt.xlabel('Initial position in asset 1')
        plt.ylabel('Initial position in asset 0')
        plt.title('Shape of the no trade region')
    
        
        plt.show()
        
    else : 
        
        fig = plt.figure()
        ax = fig.add_subplot()
        plt.scatter(abscisses,ordonnees)
        plt.plot(line_1_x_values,line_1_y_values,color='black')
        plt.plot(line_2_x_values,line_2_y_values,color='black')
        plt.plot(line_3_x_values,line_3_y_values,color='black')
        plt.plot(line_4_x_values,line_4_y_values,color='black')
        plt.plot(boundary_1_x_values,boundary_1_y_values,color='black')
        plt.plot(boundary_2_x_values,boundary_2_y_values,color='black')
        plt.plot(boundary_3_x_values,boundary_3_y_values,color='black')
        plt.plot(boundary_4_x_values,boundary_4_y_values,color='black')
        plt.plot(boundary_5_x_values,boundary_5_y_values,color='black')
        plt.plot(boundary_6_x_values,boundary_6_y_values,color='black')
        plt.plot(boundary_7_x_values,boundary_7_y_values,color='black')
        plt.plot(boundary_8_x_values,boundary_8_y_values,color='black')
    


        plt.xlabel('Initial position in asset 1')
        plt.ylabel('Initial position in asset 0')
        plt.title('Shape of the no trade region')
    
        
        plt.show()
        
    
    
    

```

We give first the position on asset 1 then on asset 2, for instance, BUY/SELL means taking a long positon on  asset 1 and a short 1 on asset 2
